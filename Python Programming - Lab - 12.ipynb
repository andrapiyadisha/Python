```python
x = range(1,11,1)
cxMarks= [8,9,6,3,5,7,4,1,2,6]
cyMarks= [5,8,9,6,3,2,4,8,8,9]
plt.plot(x,cxMarks, ls=":", lw=1.5, c="r", marker="<",label = "CX")
plt.plot(x, cyMarks, ls="--", lw=1.5 , c="b", marker="o", label = "CY")
plt.xlabel("Roll No")
plt.ylabel("CPI")
plt.text(x[cxMarks.index(min(cxMarks))],min(cxMarks),"Lowest CPI")
plt.legend()
plt.show()
```

### 06) WAP to demonstrate the use of Scatter Plot.

```python
import random
a = [random.random() for i in range(100)]
b = [random.random() for i in range(100)]
plt.scatter(a,b,c=b, cmap="plasma")
plt.colorbar()
plt.show()
```

### 07) WAP to demonstrate the use of Histogram.

```python
import random
a = [random.random() for i in range(100)]
plt.hist(a,bins=6,histtype="bar")
plt.show()
```

### 08) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
a = ["a","b","c","d","e"]
b = [1000,2000,5000,2200,4000]
bars = plt.bar(a,b)
for i in bars:
    yc = i.get_height()
    plt.text(i.get_x(),yc+100,f"{yc}")
plt.show()
```

### 09) WAP create a Scatter Plot with several colors in Matplotlib?

```python
import random
a = [random.random() for i in range(100) if i%2 == 0]
b = [random.random() for i in range(100) if i%2 == 0]
plt.scatter(a,b,c=b, cmap="plasma")
plt.colorbar()
plt.show()
```
